```python
import pandas as pd
import numpy as np
```

Reading Dataset

```python
data=pd.read_csv('/content/Iphone 15 Pro Max- Buy Products Online at Best Price in India - All Categories _ Flipkart.com.csv')
data.head(10)
```

Duplicate Value Removal

```python
duplicate_rows=data[data.duplicated()]
num_duplicates=len(duplicate_rows)
print("Number of Duplicate Rows: -" + str(num_duplicates))
```

```python
df_without_duplicate=data.drop_duplicates()
df_without_duplicate.to_csv('NewFile.csv',index=False)
```

```python
data1=pd.read_csv("/content/NewFile.csv")
data1.head(10)
```

```python
duplicate_rows=data1[data1.duplicated()]
num_duplicates=len(duplicate_rows)
print("Number of Duplicate Rows: -" + str(num_duplicates))
```

Removal of NaN Values

```python
data1.dropna(inplace=True)
```

```python
data1.head(10)
```

Text Cleaning

```python
data1['Ratings']=data1['Ratings'].str.replace(' Ratings', '')
data1.head(10)
```

```python
data1['Reviews']=data1['Reviews'].str.replace('Reviews', '')
data1.head(10)
```

#DATA VISUALIZATION

```python
summary_stats = data1.describe()
print(summary_stats)
```

```python
import matplotlib.pyplot as plt

plt.hist(data1['Price'], bins=10)
plt.xlabel('Price')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Frequency')
plt.title('Distribution of iPhone Prices')
plt.show()
```

```python
data1['Price'] = data1['Price'].str.replace('₹', '')
data1['Price'] = data1['Price'].str.replace(',', '')
data1['Price'] = pd.to_numeric(data1['Price'])
plt.figure(figsize=(10, 6))
plt.boxplot(data1['Price'])
plt.ylabel('Price (in ₹)')
plt.title('Boxplot of iPhone Prices')
plt.show()
```

```python
plt.hist(data1['Ratings'], bins=10, alpha=0.5, label='Ratings')
plt.hist(data1['Reviews'], bins=10, alpha=0.5, label='Reviews')
plt.xlabel('Count')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings and Reviews')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data1['Name'], data1['Price'])
plt.xlabel('iPhone Models')
plt.ylabel('Price (in ₹)')
plt.title('Comparison of iPhone Prices')
plt.xticks(rotation=90, ha='right')
plt.show()
```
